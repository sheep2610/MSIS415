```python
my_file = "UMB_Code_of_Conduct.txt"
with open(my_file, "r", encoding="ISO-8859-1") as file:  #open and read file 
    text = file.read()
print(text)
```

```python
# Convert the text to lowercase
text = text.lower()
print(text)
```

```python
import string                              
word_list = []                              # split based on spaces
for e in text.split() :
    e = e.replace('\n', ' ')
    while len(e) > 0 and e[-1] in string.punctuation:
        e = e[:-1]
    while len(e) > 0 and e[0] in string.punctuation:
        e = e[1:]
    if e:
        word_list.append(e)
print(word_list)
```

```python
count_dict = {}                 # initialization 
for word in word_list: 
    if word in count_dict:
        count_dict[word] += 1   # increase count
    else:
        count_dict[word] = 1    # add new word
print(count_dict)  
```

#### question 2
#####  Use dictionary to represent the frequencies. Words are keys of the dictionary, and word counts are values

```python
file = open("result.txt", "w")                  # open file to write
for key, value in count_dict.items():           # iterate dictionary
    print(str(key) + ":" + str(value), file=file)  
file.close()                                    
with open("result.txt", "r") as file:
    print(file.read())                          # read and print file 
```
