```python
my_file = "UMB_Code_of_Conduct.txt"
with open(my_file, "r", encoding="ISO-8859-1") as file:  #open and read file 
    text = file.read()
print(text)
```

```python
import string
count_dict = {}  # Dictionary to store word frequency
for e in text.split(' '):
    e = e.replace('\n', '')  
    while len(e) > 0 and e[-1] in string.punctuation:  
        e = e[:-1]
    while len(e) > 0 and e[0] in string.punctuation:
        e = e[1:]   
    print(e) 
```
