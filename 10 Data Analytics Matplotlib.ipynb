```python
import matplotlib.pyplot as plt
import numpy as np

#createing data points
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

#preparing figure
plt.figure(figsize = (12,6))
ax = plt.axes()

ax.plot(x,y)
ax.plot(x, np.zeros(x.shape), '--')
```

```python
import matplotlib.pyplot as plt
import numpy as np

#createing data points
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
z = np.cos(x)

#preparing figure
plt.figure(figsize = (12,6))
ax = plt.axes()

ax.plot(x, y, color='blue', label='Sine wave')
ax.plot(x, z, color='black', label='Cosine wave')
ax.plot(x, np.zeros(x.shape), '--', color='red')
ax.set_xlim(0, 2*np.pi)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#createing data points
x = np.linspace(-10, 10, 100)
y = x*x

#preparing figure
plt.figure(figsize = (4,3))
ax = plt.axes()
ax.plot(x,y)
ax.plot(np.zeros(x.shape), np.linspace(0, 100, 100), '--')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#createing data points
x = np.linspace(-3, 3, 100)
y = 1/math.sqrt(2 * math.pi) * math.e ** (-0.5*x*x)

#preparing figure
plt.figure(figsize = (6,3))
ax = plt.axes()

ax.plot(x,y)
ax.plot(np.zeros(y.shape), np.linspace(0.0, 0.4, 100), '--')
plt.show()
```
