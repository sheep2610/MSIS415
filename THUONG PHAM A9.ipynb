##### 4) (2point) Display age and payment method data for the last 15 customers

```python
cus_data = pd.read_csv('Customer.csv', sep=',', header=0)
cus_data.loc[cus_data.index[-15:], ['Age', 'Payment Method']]
```

##### 5) (2point) Display age and payment method data for the youngest 12 customers
