##### 1: (2 points) First, create dummies for features: Sex and Embarked. Then put them back to the df. Select the appropriate features (remember to exclude PassengerID, which is just the index column) to build your trainX, trainY, testX, and testY. Use df.iloc[:600] for training set, df.iloc[600:] for test set.

```python
# create dummies for features: Sex and Embarked
sex_dummies = pd.get_dummies(df["Sex"])
embarked_dummies = pd.get_dummies(df["Embarked"])
df = df.join([sex_dummies, embarked_dummies])
```

```python
#Select the appropriate features
trainX = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'female', 'male', 'C', 'Q', 'S']][:600]
trainY = df[['Survived']][:600]

testX = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'female', 'male', 'C', 'Q', 'S']][600:]
testY = df[['Survived']][600:]
```

#####  (4 points) Use sklearn’s DecisionTreeClassifier and svm.SVC to build two classifier predicting y Survived using X ‐> [Pclass, Sex, Age, SibSp, Parch, Fare, Embarked]. Save the two models in two different python variables.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(trainX, trainY)

svm_model = SVC()
svm_model.fit(trainX, trainY)
```

##### 3: (2 points) For each model, calculate its accuracy on the test dataset by calling the odel.predict(testX) function. Use sklearn.metrics.accuracy_score(y_true, y_pred) to calculate the score. (1 points) Display the two model’s scores in one cell to compare the models. (1 points) Write a one sentence summary about your findings and decisions on which model to use.

```python
from sklearn.metrics import accuracy_score
```

```python
# call the model.predict(testX) function
tree_pred = tree_model.predict(testX)
svm_pred = svm_model.predict(testX)

# use accuracy_score(y_true, y_pred) to calculate the score
tree_acc = accuracy_score(testY, tree_pred)
svm_acc = accuracy_score(testY, svm_pred)

# Display both scores in one cell
print("Decision Tree Score:", tree_acc)
print("SVM Score:", svm_acc)
```

#### Decision Tree Score gave higher accuracy, so we will use it for the prediction
